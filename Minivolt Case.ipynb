```python
from solid import *
from solid.utils import *
```

```python
twiddle = 0.001

def meter():

    thick_ear_padding = 1.0
    ear_cutting = True
    ear_width_actual = 33.05  # From outside of both
    ear_width_pad = 1
    ear_width = ear_width_actual + ear_width_pad

    ear_stick_actual = 5.27  # How far the ears stick out
    ear_stick = ear_stick_actual - (ear_width_pad/2)
    
    ear_spacing = 4.94  # How far up the body the ears start
    ear_height = 4.95  # How tall the ears alone are
    ear_thickness = 0.94
    if ear_cutting:
        ear_height += ear_spacing
    ears = cube(size=[ear_width, ear_thickness + thick_ear_padding, ear_height])
    placed_ears = translate([
            -ear_stick,
            0-(thick_ear_padding/2),
            ear_spacing
        ])(ears)

    # Body is the meter and chips on the back
    body_back_pad = 0.5
    body_depth_actual = 9.93
    body_depth = body_depth_actual + body_back_pad
    body_height = 14.83
    led_depth = 7.4
    led_width_actual = 22.57
    led_width_pad = 1
    led_width = led_width_actual + led_width_pad
    body = cube([led_width, body_depth, body_height])
    placed_body = translate([-(led_width_pad/2),-led_depth,0])(body)
    
    meter = color(Blue)(placed_ears + placed_body)

    box_wall = 2.5
    box_wall2 = box_wall * 2
    back_half_depth = body_depth - (led_depth + (ear_thickness /2))
    box_depth = (back_half_depth+box_wall)*2
    box = cube([
            ear_width+box_wall2,
            box_depth,
            body_height+box_wall
        ])
    
    box_y_offset = -(box_wall+(thick_ear_padding/2)+(ear_thickness/2))
    placed_box = translate([
            -(ear_stick+box_wall),
            box_y_offset,
            # Twiddle the placement to make OpenSCAD preview not suck
            -(box_wall+twiddle)
        ])(box)

    base = color(Red)(cube([
            ear_width+box_wall2,
            box_depth,
            box_wall
        ]))

    placed_base = translate([
            -(ear_stick+box_wall),
            box_y_offset,
            # Twiddle the placement to make OpenSCAD preview not suck
            body_height+box_wall
        ])(base)
    
    #hole = translate([12.75,box_y_offset+box_wall,-0.001])(
    #    cylinder(h=200, r=2.5))
    hole = cylinder(h=200, r=2.5)
    hole += translate([-2.5,-box_wall2,0])(cube([5,box_wall2,200]))
    hole = translate([12.75,box_y_offset+box_wall,-0.001])(hole)
    base_sliced = placed_base - hole

    return {'box': placed_box - meter, 'meter': meter, 'base': base_sliced}

scad = '$fn=16;\n'
scad += scad_render(meter()['box'])
scad += scad_render(meter()['base'])
open('minivolt.scad', 'w').write(scad)
print scad
```
